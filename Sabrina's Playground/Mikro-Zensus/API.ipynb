```python
def find_one_cat(string, category='all', pagelength='16', language='de'):

    # Set url
    url = 'https://www-genesis.destatis.de/genesisWS/rest/2020/find/find?'

    headers = {'username': os.getenv('user'),
            'password': os.getenv('password')}
    
    parameters = {'term': {string},
                 'category': {category},
                 'pagelength': {pagelength},
                 'language': {language}}
    
    # Request data from url
    r = requests.get(url, headers=headers, params=parameters)
    
    # Decode response with json decoder
    find_temp = r.json()

    # Flatten json response
    find_df = pd.json_normalize(find_temp,
                                     sep = '_',
                                     record_path = 'Statistics')        # All, Cubes, Statistics, Tables, Timeseries, Variables
    
    # Print final dataset
    return find_df
    #return find_temp
```

```python
find_one_cat('Abfall', pagelength='20')
```
